```python
#Visualize the distribution of Engagement_Score across different Campaign_Type using a box plot.
```

```python
Engagement_Score=df_csv.Engagement_Score
```

```python
plt.figure(figsize=(6, 5))
plt.boxplot(Engagement_Score,
            patch_artist=True,
            boxprops    = dict(facecolor='lightgreen',color='black'),
            medianprops = dict(color='red'),
            whiskerprops= dict(color='yellow'),
            capprops    = dict(color='red'),
            flierprops  = dict(marker='o',markerfacecolor='orange'))


plt.title("Box Plot of Python Scores")
plt.ylabel("Campaign_Type")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Customer Segmentation:

Create a count plot to visualize the distribution of Target_Audience.
Identify which Customer_Segment has the highest Conversion_Rate for each
Language using a bar chart.
Visualize the distribution of Acquisition_Cost across each Customer_Segment,
categorized by Channel_Used, using a box plot.
Analyze average Conversion_Rate by Language using a bar chart to compare
the effectiveness of campaigns conducted in different languages.
```

```python
#Create a count plot to visualize the distribution of Target_Audience.
```

```python
sns.countplot(
    x="Location",
    hue="Target_Audience",
    data=df_csv,
    palette="Set3",
    order= ["Miami","Chicago","New York","Houston"]
)
plt.title("Distribution of Target Audience")
plt.legend(bbox_to_anchor=(1,1)) # 90% (1,1)
plt.show()
```

```python
#Identify which Customer_Segment has the highest Conversion_Rate for each
Language using a bar chart.
```

```python
df_csv.head()
```

```python
import matplotlib.pyplot as plt
```

```python
High = df_csv.groupby(['Language', 'Customer_Segment'])['Conversion_Rate'].mean().reset_index()
```

```python
top_segments = High.loc[High.groupby('Language')['Conversion_Rate'].idxmax()]
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=top_segments, x='Language', y='Conversion_Rate', hue='Customer_Segment', palette='Set1')


plt.title('Customer_Segment for Each Language', fontsize=14)
plt.xlabel('Language', fontsize=12)
plt.ylabel('High', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Visualize the distribution of Acquisition_Cost across each Customer_Segment,
categorized by Channel_Used, using a box plot.
```

```python
sns.boxplot(
    x="Customer_Segment",
    y="Acquisition_Cost",
    hue="Channel_Used",
    data=df_csv,
    showfliers=True,
    palette="bright",
    orient="v"
)
plt.title("Distribution of Acquisiton cost across each Customer segment")
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```
